```python
import cv2
from scipy.spatial import distance
import numpy as np

#function for recognizing an expression by an image
def recognize_expression(img):

    expression_label = "joy" # label happy

    return expression_label


expression_database = {
    "joy": cv2.imread(uploaded_image_path),
    "sadness": cv2.imread(uploaded_image_path_sad), #I made the database this way to be expanded in the future with a larger dataset of images
}

# comparing two images
def compare(image1, image2):
    # image to greyscale
    greyOne = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
    greyTwo = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

    # resize images
    resizedImage1 = cv2.resize(greyOne, (100, 100))
    resizedImage2 = cv2.resize(greyTwo, (100, 100))

    #parameters to make HOGDescriptor work:
    winSize = (32,32)
    blockSize = (16,16)
    blockStride = (8,8)
    cellSize = (8,8)
    nbins = 9
    hog = cv2.HOGDescriptor(winSize, blockSize, blockStride, cellSize, nbins)

    #Two hog variables
    hogOne = hog.compute(greyOne)
    hogTwo = hog.compute(greyTwo)

    # compare hog distances of difference using euclidean function
    dist = distance.euclidean(hogOne, hogTwo)

    #return distances
    return dist

# find closest expression by image
def closestExpression(img):
    # call recognize expression function
    uploaded_expression = recognize_expression(img)

    # compare each image for closest match
    closest_match = None
    min_distance = float('inf')
    for expression, expression_image in expression_database.items():
        similarity = compare(img, expression_image)
        if similarity < min_distance:
            min_distance = similarity
            closest_match = expression

    return closest_match

# process image
def process_uploaded_image(img):
    # closest image
    closest_expression = closestExpression(img)

    print("Nearest Match:", closest_expression)



# Main Section

# test image upload
uploaded = files.upload()


#get file name
filename = next(iter(uploaded))



# save file
for filename in uploaded.keys():
    uploaded_image_path = filename

print(filename)

# read uploaded image
uploaded_image = cv2.imread(filename)

# call function to process file
process_uploaded_image(uploaded_image)
```
